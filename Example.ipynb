```python
# importing dexter and pandas

import dexter as dxt
import pandas as pd
```

```python
# the most used class in dexter is the FrameMap

df1 = pd.DataFrame({'First': [1, 2, 3], 'second': [4, 5, 6]})
df2 = pd.DataFrame({'Third': [7, 8, 9], 'fourth': [10, 11, 12]})

# you can construct a FrameMap from a list of dataframes and a list of their names

dataframes = dxt.FrameMap([df1, df2], ['df1', 'df2'])

# the display method is used to visualize a FrameMap

dataframes.display()
```

```python
# alternatively you can create a FrameMap by reading multiple files in a folder with readm functions
dataframes = dxt.readm_csv('./usecases/movielens/ml-25m/', chunksize = 20000)
```

```python
# you can easily access the names and dataframes used to create the FrameMap

dataframes.frames
dataframes.names
```

```python
# as we can see,

print('ratings memory usage reduced to {:.2f}%'.format((400001648 / 400001648) * 100))
print('movies memory usage reduced to {:.2f}%'.format((5887991 / 10200589) * 100))
print('genome-scores memory usage reduced to {:.2f}%'.format((155844608 / 374026880) * 100))
print('genome-tags memory usage reduced to {:.2f}%'.format((77983 / 84751) * 100))
print('links memory usage reduced to {:.2f}%'.format((749204 / 1498280) * 100))
print('tags memory usage reduced to {:.2f}%'.format((24822779 / 100284123) * 100), end='\n\n')

print('total memory usage reduced to {:.2f}%'.format((400001648 + 5887991 + 155844608 + 77983 + 749204 + 24822779) /
                                                (400001648 + 10200589 + 374026880 + 84751 + 1498280 + 100284123) * 100))
```
